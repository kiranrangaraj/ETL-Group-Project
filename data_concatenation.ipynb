```python
# Import dependency
import pandas as pd
```

```python
# Load both of the cleaned-up CSV data files
tweets_1 = 'Output_data/tweets_1.csv'
tweets_2 = 'Output_data/tweets_2.csv'

# Read the loaded files in a Pandas dataframe
df_1 = pd.read_csv(tweets_1)
df_2 = pd.read_csv(tweets_2)
```

```python
# Concatenate both of the dataframes
frames = [df_1, df_2]
result = pd.concat(frames)
```

```python
# Reset the index
concat = result.reset_index(drop=True)

# Specify desired columns in dataframe
concat_df = concat[["Tweet_Id", "Date", "Time", "Text", "Retweets", "Favorites"]]

# Set the index to the date to serve as a primary key in the SQL database
tweet_df = concat_df.set_index('Tweet_Id')

# Display the concatenated dataframe
tweet_df
```

```python
# Save the Trump-tweets dataframe as a new CSV file to be imported into the SQL database
tweet_df.to_csv('Output_data/2016-2020_tweets.csv')
```
