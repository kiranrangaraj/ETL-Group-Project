```python
# Import all dependencies
import pandas as pd
```

```python
# Load both of the CSV data files
tweets_1 = 'Output_data/tweets_1.csv'
tweets_2 = 'Output_data/tweets_2.csv'

# Read the loaded files in a Pandas dataframe
df_1 = pd.read_csv(tweets_1)
df_2 = pd.read_csv(tweets_2)
```

```python
# Concatenate both of the dataframes
frames = [df_1, df_2]
result = pd.concat(frames)
```

```python
# Reset the index
concat = result.reset_index(drop=True)

# Specify desired columns in dataframe
concat_df = concat[["Datetime", "Text", "Tweet_Id", "Retweets", "Favorites"]]

# Display the concatenated dataframe
concat_df
```

```python
# Save to a new csv file
concat_df.to_csv('Output_data/2016-2020_tweets.csv')
```
