```python
# Import all dependencies
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the csv data files
trump_tweets = 'Output_data/realDonaldTrump-tweets.csv'

# Read the csv file as a Pandas dataframe
tweets_df = pd.read_csv(trump_tweets)

# Show the dataframe
tweets_df
```

```python
# Keep all of rows of tweets that were not replies to someone else, ie. tweets Trump made to his main Twitter page
tweets_df = tweets_df[pd.isnull(tweets_df['Reply to'])]

# Remove the "Reply to" column
trump_tweets = tweets_df[["Datetime", "Text", "Tweet Id", "Retweets", "Favorites"]]

# Rename 'Tweet Id' column to be consistent with pre-election column name
trump_tweets = trump_tweets.rename(columns={"Tweet Id":"Tweet_Id"})

# Drop any row that contains null values
trump_tweets.dropna(inplace=True)

# Reset the index
tweets = trump_tweets.reset_index(drop=True)

# Preview the cleaned-up dataframe
tweets
```

```python
# Save to a new csv file
tweets.to_csv('Output_data/tweets_1.csv')
```
