```python
# Import Dependencies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
import json
%matplotlib inline
```

```python
# Load JSON file containing 2016 Trump Tweets
tweets = pd.read_json("Resources/2016tweets.json")
tweets.head()
```

```python
# Get names of indexes for which column "is_retweet" is True
indexNames = tweets[tweets['is_retweet'] == True].index

# Delete these row indexes from dataFrame
tweets.drop(indexNames , inplace=True)
```

```python
# Drop the "source" & "is_retweet" columns
tweets1 = tweets.drop(columns=['source','is_retweet'])
```

```python
# Rename the columns to match the column names of other dataframes being used in this project
tweets2 = tweets1.rename(columns={"text":"Text","created_at":"Datetime", "retweet_count" :"Retweets",
                                  "id_str":"Tweet_Id","favorite_count":"Favorites"})
# Preview dataframe
tweets2.head()
```

```python
# Reorder the columns to match the column order of other dataframes to be merged with
tweets3 = tweets2[["Datetime", "Text", "Tweet_Id", "Retweets", "Favorites"]]

# Sort the table according to "Datetime" in descending order
tweets3.sort_values(by=['Datetime'], inplace=True, ascending=False)

# Reset the index
tweets4 = tweets3.reset_index(drop=True)

# Remove timezone information from datetime format
tweets4['Datetime'] = tweets4['Datetime'].apply(lambda x: x.replace(tzinfo=None))

# Show preview of cleaned-up dataframe
tweets4
```

```python
# Save to a new csv file
tweets4.to_csv('Output_data/tweets_2.csv')
```
